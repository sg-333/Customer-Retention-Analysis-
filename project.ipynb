### Data Cleaning and Transformation

Removing unnecessary columns:

```python
cols = [
    'CLIENTNUM', 'Education_Level', 'Marital_Status', 'Avg_Open_To_Buy', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'
]

for col in cols:
    df.drop([col], axis=1, inplace=True)
```

##### Outlier Detection

```python
# Checking the min, mean, median, max values of numerical columns
num_cols = df.select_dtypes(include=['number']).columns
summary = pd.DataFrame({
    'Min': df[num_cols].min(),
    'Mean': df[num_cols].mean(),
    'Median': df[num_cols].median(),
    'Max': df[num_cols].max()
})

summary = summary.applymap(lambda x: f"{x:.1f}")
summary
```

```python
# Displaying box plot of all extreme outliers seen above
out_cols = ['Credit_Limit', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt', 'Total_Ct_Chng_Q4_Q1']
for col in out_cols:
    plt.figure(figsize=[10,1])
    sns.boxplot(x=df[col])
    plt.show()
```

##### Feature Engineering:

1. Transaction Frequency

```python
# Creating new feature for getting the transaction frequency of the customer
df['Avg_Transactions_per_Month'] = df['Total_Trans_Ct'] / df['Months_on_book']
df.sample()
```

2. Inactivity Ratio:

```python
# Creating new feature for getting the inactivity of the customer
df['Inactivity_ratio'] = df['Months_Inactive_12_mon'] / 12
df.sample()
```
